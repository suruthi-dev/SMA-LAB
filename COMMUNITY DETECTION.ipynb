```python
import networkx as nx
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings(action='ignore')
```

```python
G = nx.barbell_graph(7,4)
```

```python
nx.draw_networkx(G)
```

# girvan_newman

```python
from networkx.algorithms.community.centrality import girvan_newman
```

```python
communities = girvan_newman(G)
```

```python
def draw_graph(community):
    node_groups = []
    for com in next(community):
        node_groups.append(list(com))
        
    # will depend on the number of groups

    color_map = []
    for node in G:
        if node in node_groups[0]:
            color_map.append('blue')
        else: 
            color_map.append('green') 
    nx.draw_networkx(G, node_color=color_map, with_labels=True)
    plt.show()
```

```python
draw_graph(communities)
```

# louvain_communities

```python
from networkx.algorithms.community import louvain_communities
```

```python
louvain_community = louvain_communities(G)
```

```python

# Convert communities to a list of lists
communities = [list(community) for community in louvain_community]
```

```python
# Visualize the communities
def draw_graph(G, communities):
    pos = nx.spring_layout(G)  # You can choose a layout algorithm that you prefer
    cmap = plt.cm.get_cmap("viridis", len(communities))
    
    for i, community_nodes in enumerate(communities):
        nx.draw_networkx_nodes(G, pos, nodelist=community_nodes, node_size=200, node_color=[cmap(i)])
        
    nx.draw_networkx_edges(G, pos, alpha=0.5)
    nx.draw_networkx_labels(G, pos)
    plt.axis("off")
    plt.show()
```

```python
draw_graph(G, communities)
```

# greedy_modularity_communities

```python
from networkx.algorithms.community import greedy_modularity_communities
```

```python
greedy = greedy_modularity_communities(G)
```

```python
# Convert communities to a list of lists
communities = [list(community) for community in greedy]
```

```python
# Visualize the communities with numbered nodes
def draw_graph_with_node_numbers(G, partition):
    pos = nx.spring_layout(G, seed=4)  # Use a fixed seed for consistent layout
    cmap = plt.cm.get_cmap("viridis", len(partition))
    
    for i, community_nodes in enumerate(partition):
        subgraph = G.subgraph(community_nodes)
        nx.draw_networkx_nodes(subgraph, pos, node_size=200, node_color=[cmap(i)], label=f'Community {i}')
        node_labels = {node: str(node) for node in community_nodes}
        nx.draw_networkx_labels(subgraph, pos, labels=node_labels)
        nx.draw_networkx_edges(subgraph, pos, alpha=0.5)
    
    plt.axis("off")
    plt.legend(loc='upper right')
    plt.show()
```

```python
draw_graph_with_node_numbers(G, communities)
```

```python
from networkx.algorithms.community import naive_greedy_modularity_communities
```

```python
naive_greedy = greedy_modularity_communities(G)
```
